```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
customers=pd.read_csv('customer_churn.csv')   
customers.head(50)
```

```python
#select the features and target:
customers = customers[['SeniorCitizen','tenure','MonthlyCharges','Churn']]
customers
```

```python
cols=[]
for c in customers.columns:
    cols.append(c.lower())
       
customers.columns = cols
customers.columns
```

```python
customers["churn"].value_counts(dropna=False)
```

```python
#Convert it to numerical
def transform_churn(x):
    if x=='Yes':
        return 1
    else:
        return 0
# customers['churn']=customers["churn"].apply(transform_churn)

# customers['churn']
customers["churn"] = list(map(transform_churn,customers["churn"]))
display(customers['churn'])
```

```python
# X/y split
X = customers.drop('churn',axis = 1)
y = customers['churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
X_train_num = X_train.select_dtypes(include = np.number)
# all features are numeric, so no need to split into _num and _cat
transformer = StandardScaler().fit(X_train)
# X_train_scaled = pd.DataFrame(transformer.transform(X_train),columns=X.columns)
X_train_scaled = transformer.transform(X_train)
# because this is the only tranformation we apply to the numerics,
# we can immediately transform the X_test as well
X_test_scaled = pd.DataFrame(transformer.transform(X_test),columns=X.columns)  
X_test_scaled


```

```python
#Linear relation

from sklearn.linear_model import LogisticRegression
classification = LogisticRegression(random_state=0, solver='saga',
                  multi_class='multinomial').fit(X_train_scaled, y_train)
```

```python
#Prediction Time!
predictions = classification.predict(X_test_scaled)
predictions
classification.score(X_test_scaled, y_test)
#with a simple model we get a high accuracy score of .79. 
```

```python
#The precision, recall and F1 score show this imbalance better in scores. The score lowered to 0.45 for recall, showing it's not a very good model
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

pred = classification.predict(X_test_scaled)

print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```

```python
Precision:Precision is 0.61, it indicates that 61% of the churn predicted as spam are indeed spam. Recall: recall tells you the percentage of actual spam emails that were correctly predicted as spam. A high recall means the model can effectively capture most spam emails. If recall is 0.90, it means the model identified 90% of the actual spam emails. F1 Score: For instance, if F1 is 0.92, it means the model achieves a balance between precision and recall.
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)   
```
